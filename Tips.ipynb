```python
#import libraries

#suporting libraries
import pandas
import numpy

#preparations
from sklearn import preprocessing
from sklearn import __version__ as sklearn__version__
from sklearn.model_selection import train_test_split

#classifiers
from sklearn.naive_bayes import GaussianNB
from sklearn import tree

#regression algorithms
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression

#clustering algorithms
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
```

```python
#ETL data

def cSex(sex):
    mapping = {'Male': 0, 'Female': 1}
    return mapping.get(sex)

def cSmoker(smoker):
    mapping = {'Yes': True, 'No': False}
    return mapping.get(smoker)

def cDay(day):
    mapping = {'Mon': 1, 'Tue': 2, 'Wed': 3, 'Thur': 4, 'Fri': 5, 'Sat': 6, 'Sun': 7}
    return mapping.get(day)

def cTime(time):
    mapping = {'Breakfast': 1, 'Lunch': 2, 'Dinner': 3}
    return mapping.get(time)

tips = pandas.read_csv('~\AnacondaProjects\\Practice\\tips.csv', converters = {'sex': cSex, 'smoker': cSmoker, 'day': cDay, 'time': cTime})
tips
```

```python
#EDA

#initial look at data
#tips

#unique values in non-continuous variables
for column in tips.columns:
    if tips[column].dtype != 'float64':
        print(column)
        print(tips[column].unique())
        print(tips[column].value_counts())
        print('\n')
        print(tips.groupby(column).mean())
    print('\n')
    print('\n')
    
#class labels' ratios
smokerCount = len(tips[tips['smoker'] == True])
nonSmokerCount = len(tips[tips['smoker'] == False])
smokerPercent = (smokerCount/(smokerCount + nonSmokerCount)) * 100
nonSmokerPercent = (nonSmokerCount/(smokerCount + nonSmokerCount)) * 100
#https://stackoverflow.com/questions/15286401/print-multiple-arguments-in-python
print('Smokers: Count = {c}, {p:.2f}%'.format(c = smokerCount, p = smokerPercent))
print('Non-Smokers: Count = {c}, {p:.2f}%'.format(c = nonSmokerCount, p = nonSmokerPercent))

#plots : TODO

#TODO: create some dummy datapints from training dataset for any class which is proportionally under-represented USE Over-sampling using SMOTE
```

```python
#splice data into test and train subsets

train,test = train_test_split(tips, test_size=0.2) #explicitly specify random state to ensure splits are the same every execution

trainX = train[['total_bill', 'tip', 'sex', 'day', 'time', 'size']]
trainY = train[['smoker']]

testX = test[['total_bill', 'tip', 'sex', 'day', 'time', 'size']]
testY = test[['smoker']]
```

```python
#PCA https://stackabuse.com/implementing-pca-in-python-with-scikit-learn/
```

```python
#supervised - classification - naive bayes - best for categorical outputs with no correlation between variables

naiveBayes = GaussianNB();
naiveBayes.fit(trainX, numpy.ravel(trainY))
predictions = naiveBayes.predict(testX)
score = (naiveBayes.score(trainX, trainY)) * 100

print('Accuracy : {:.2F}%'.format(score))
```

```python
#supervised - classification - decision tree - best for categorical outputs

decisionTree = tree.DecisionTreeClassifier()
decisionTree.fit(trainX, trainY)
predictions = decisionTree.predict(testX)
score = (decisionTree.score(trainX, trainY)) * 100

print('Accuracy : {:.2F}%'.format(score))
```

```python
#supervised - regression - logistic - best for continuous outputs
#https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html
#https://engineering.eckovation.com/logistic-regression-introduction-implementation-using-sklearn/

logisticRegression = LogisticRegression()
logisticRegression.fit(trainX, numpy.ravel(trainY))
predictions = logisticRegression.predict(testX)
score = (logisticRegression.score(testX, testY)) * 100

print(predictions)
print(testY)
print('Accuracy : {:.2F}%'.format(score))
```

```python
#supervised - regression - linear - best for continuous outputs
#https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

linearRegression = LinearRegression()
linearRegression.fit(trainX, numpy.ravel(trainY))
predictions = linearRegression.predict(testX)
score = (linearRegression.score(testX, testY)) * 100

print(predictions)
print(testY)
print('Accuracy : {:.2F}%'.format(score))
```

```python
#preparation for unsupervised learning

tipsX = tips[['total_bill', 'tip', 'sex', 'day', 'time', 'size']]
tipsY = tips[['smoker']]
```

```python
#unsupervised - clustering - K-Means
#https://scikit-learn.org/stable/modules/clustering.html#k-means
#https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html

kMeansClustering = KMeans(n_clusters = 2, max_iter = 3000)
predictions = kMeansClustering.fit_predict(tipsX)
print(predictions)
print(numpy.ravel(tipsY))
```

```python
#unsupervised - clustering - Agglomerative
#https://scikit-learn.org/stable/modules/clustering.html#hierarchical-clustering
#https://scikit-learn.org/stable/modules/generated/sklearn.cluster.AgglomerativeClustering.html

agglomerativeClustering = AgglomerativeClustering(n_clusters = 2)
predictions = agglomerativeClustering.fit_predict(tipsX)
print(predictions)
print(numpy.ravel(tipsY))
```

```python
#unsupervised - clustering - DBSCAN
#https://scikit-learn.org/stable/modules/clustering.html#dbscan
#https://scikit-learn.org/stable/modules/generated/sklearn.cluster.DBSCAN.html

dbscan = DBSCAN(eps = 3, min_samples = 7)
predictions = dbscan.fit_predict(tipsX)
print(predictions)
print(numpy.ravel(tipsY))
```
